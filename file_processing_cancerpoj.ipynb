```python
import pandas as pd
res = pd.read_csv('output/gene_names.csv')
wp = pd.read_csv('output/ferroptosis_wp.csv')
ferrdb = pd.read_csv('output/ferrdb_filtered.csv')
pyro = pd.read_csv('output/pyro_genes.csv')
necro = pd.read_csv('output/necro_genes.csv')
```

```python
res["wp"] = res.external_gene_name.isin(wp.Name)
res["ferrdb"] = res.external_gene_name.isin(ferrdb.Name)
ferr_genes = res.loc[(res['wp'] == True)  | (res['ferrdb'] == True)]
```

```python
def highlight_rows(row):
    if row["wp"] == True and row["ferrdb"] == False :
        return ['background-color: yellow'] * len(row)
    elif row["wp"] == False and row["ferrdb"] == True:
        return ['background-color: lightblue'] * len(row)
    elif row["wp"] == True and row["ferrdb"] == True:
        return ['background-color: lightgreen'] * len(row)
    else :
        return ['background-color: red'] * len(row)
    
fg = ferr_genes.style.apply(highlight_rows, axis=1)
fg
```

```python
fg.to_excel("output/ferroptosis_genes_filtered_2.xlsx")
```

```python
#pyroptosis genes
res["pyro"] = res.external_gene_name.isin(pyro.Name)
pyro_genes = res.loc[res['pyro'] == True]
pyro_genes.to_excel("output/pyroptosis_genes.xlsx")
print(len(pyro_genes))
```

```python
#necroptosis genes
res["necro"] = res.external_gene_name.isin(necro.Name)
necro_genes = res.loc[res['necro'] == True]
necro_genes.to_excel("output/necroptosis_genes.xlsx")
```
